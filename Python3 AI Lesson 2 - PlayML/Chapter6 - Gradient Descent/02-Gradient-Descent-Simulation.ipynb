```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 绘制需要的x的点(x=[-1, 6]，共140个点)
plot_x = np.linspace(-1, 6, 141)
# 损失函数
plot_y = (plot_x - 2.5)**2 - 1
```

```python
plt.plot(plot_x, plot_y)
plt.show()
```

```python
# 对损失函数 J 求导
def dJ(theta):
    return 2 * (theta - 2.5)
```

```python
# 损失函数J
def J(theta):
    return (theta - 2.5)**2 - 1
```

```python
# eta - 学习率
eta = 0.1
# 最小变化范围，若小于episilon则不用继续了
epsilon = 1e-8

theta = 0.0

while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    
    # 判断变化值
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

print(theta)
print(J(theta))
```

```python
# 可视化每次迭代的theta
theta = 0.0
theta_history = [theta]

while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    
    # 判断变化值
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

print(len(theta_history))
plt.plot(plot_x, J(plot_x))
# 可以看到它慢慢往下迭代，直到极值位置
plt.scatter(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+')
plt.show()
```

```python
def gradient_descent(initial_theta, eta, epsilon=1e-8):
    theta = initial_theta
    theta_history.append(initial_theta)
    
    while True:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta*dJ(theta)
        theta_history.append(theta)
        
        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
    
def plot_theta_history():
    print("The count of theta caculation is: ", len(theta_history))
    plt.plot(plot_x, plot_y)
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+')
    plt.show()
```

```python
eta = 0.01
theta_history = []
gradient_descent(0., eta)
plot_theta_history()
```

```python
# 优化损失函数J
def J(theta):
    try:
        return (theta - 2.5)**2 - 1
    except:
        return float('inf')
```

```python
# 优化梯度下降函数，防止死循环
# n_iters：最多循环次数
def gradient_descent(initial_theta, eta, n_iters = 1e4, epsilon=1e-8):
    theta = initial_theta
    theta_history.append(initial_theta)
    i_iter = 0
    
    while i_iter < n_iters:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta*dJ(theta)
        theta_history.append(theta)
        
        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
        
        i_iter += 1
```

```python
eta = 1.1 # 如果不优化，这里会报错或者死循环
theta_history = []
gradient_descent(0., eta)
print(len(theta_history))
print(theta_history[-1])
```

```python
# 可视化 eta 设置过大的情境
eta = 1.2
theta_history = []
gradient_descent(0., eta, n_iters=0)
# 可以看到反向迭代过度
plot_theta_history()
```
